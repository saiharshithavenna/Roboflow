Moving all the predicted images into newly created folder

```python
import os
import shutil

# Source directory containing the images
source_directory = '/content/runs/segment/predict/'

# Destination directory (existing "predicted_image" folder)
destination_directory = '/content/runs/segment/predict/predicted_images'

# Create the destination directory if it doesn't exist
if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)


file_list = os.listdir(source_directory)


image_files = [file for file in file_list if file.endswith('.jpg')]

# Iterate through the image files and copy/move them to the destination directory
for file_name in image_files:
    source_file = os.path.join(source_directory, file_name)
    destination_file = os.path.join(destination_directory, file_name)

    # Copy or move the file from source to destination
    shutil.copy(source_file, destination_file)  # Change to shutil.move() if you want to move the files
```

Deleting the images which are mived previously

```python
import os

# Directory containing the images to delete
directory_path = '/content/runs/segment/predict/'

# List all files in the directory
file_list = os.listdir(directory_path)

# Iterate through the files and delete them
for file_name in file_list:
    file_path = os.path.join(directory_path, file_name)

    # Check if the path is a file before attempting deletion
    if os.path.isfile(file_path):
        os.remove(file_path)
```
